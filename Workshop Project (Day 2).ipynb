```python
#Importing required libraries
import rioxarray as rio
import cartopy.crs as crs
import geopandas as gpd
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.colors as Normalise
import matplotlib.cm as ScalarMappable
import rasterio
import rasterio.features
```

```python
#Loading satellite data-image
xds = rio.open_rasterio(r'C:\Users\OHIOPC\Documents\project\project_data\shp\output_SRTMGL1.tif')
xds[0].plot.imshow(vmin=0,vmax=5000,cmap='terrain')
```

```python
#Loading dataset
gdf = gpd.read_file(r'C:\Users\OHIOPC\Documents\project\project_data\shp')
gdf
```

```python
h = gdf[gdf['NAME_1'] == 'Haryana'].reset_index(drop=True)
h
```

```python
hd = list(h['NAME_2'])
hd
```

```python
#Creating Dataframe for required data to be plotted.
h_div = h.loc[h['NAME_2'].isin(hd)]
h_div.reset_index(drop=True,inplace=True)
h_div
```

```python
geometry = h_div['geometry']

#Creating a Mask to plot using geometry data
mask = rasterio.features.geometry_mask(geometry, out_shape = xds.shape[-2:], transform=xds.rio.transform(), invert=True)
masked_xds = xds.where(mask)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
img = masked_xds[0].plot.imshow(vmin=0, vmax=1500, cmap='terrain', ax=ax, cbar_kwargs = {'orientation': 'vertical','label' : 'Mean Sea level Height (in m)'})
ax.set_xlabel = ('Longitudes due East')
ax.set_ylabel = ('Latitude due North')

#Marking Location of college
h_div.plot(ec='k', fc = 'none', alpha=0.4, marker='--', ax=ax)
x, y = 76.768688, 28.450978
plt.plot(x,y, 'ok', markersize=5)
plt.text(x+0.04,y+0.01, 'College', fontsize=10,color= 'red' );

for idx, row in h_div.iterrows():
  centroid = row['geometry'].centroid
  ax.text(centroid.x+0.07, centroid.y+0.07, row['NAME_2'], ha = 'center', va = 'center', fontsize=6, color='black')

#Plotting Map
ax.set_title("Map of Haryana with college's location marked")
ax.set_aspect('auto')
plt.tight_layout()
plt.show() 
```
