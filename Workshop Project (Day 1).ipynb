# Aim: To plot the Location of college on the map using python.
   * Libraries/modules Used:
       * Numpuy
       * Matplotlib.pyplot
       * mpl_tookits.basemap

```python
# Importing required libraries.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
# Creating variable to define the size and resolution of figure to be Plotted
fig = plt.figure(figsize = (8,8), dpi=100)

# Creating object for basemap class with different parameters to define the map
m = Basemap(projection = 'lcc', 
            resolution='h',
            width=7E6, height=7E6,
            lat_0=21.43,lon_0=87.01)

m.etopo(scale=0.5, alpha=0.5)
m.bluemarble(scale=0.5)

# 28.450978, 76.768688 location of our college
x, y = m(76.768688, 28.450978)

# Plotting/ marking the position
plt.plot(x,y, 'ok', markersize=5)
plt.text(x,y, 'My College', fontsize=12,color='white')

# For Saving Figure in Desktop pc
plt.savefig('mycollege.png', dpi=600,bbox_inches='tight')
plt.show()
```
