# AIM: To visualise the eliptical orbit of earth around the sun
* Libraries/Modules Used
    * Numpy
    * Pandas
    * Matplotlib.pyplot

```python
#Importing Required Libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
n = 500
#generating angular data
t = np.linspace(0,2*np.pi,n)
x = 5*np.cos(t) + 0.4*np.random.normal(size=n)+1
y = 4*np.sin(t) + 0.4*np.random.normal(size=n)

# plotting the angular data points
plt.plot(x,y,".g")
```

```python
xmean , ymean = x.mean(), y.mean()
x -= xmean
y -= ymean

u,s,v = np.linalg.svd(np.stack((x,y)))
tt = np.linspace(0,2*np.pi,1000)

# creating eliptical path for earth to be plotted
circle = np.stack((np.cos(tt),np.sin(tt)))
transform = np.sqrt(2/n)*u.dot(np.diag(s))
fit = transform.dot(circle)+np.array(([xmean],[ymean]))
```

```python
# plotting the path
plt.plot(fit[0,:],fit[1,:],'--b')

# plotting the sun and earth's position
plt.plot(0,0,'or',markersize = 40)
plt.plot(-3.2,-2,'ob',markersize = 15)

plt.title("earth's eliptical orbit around sun")
plt.xlabel('ecix')
plt.ylabel('eciy')
```
